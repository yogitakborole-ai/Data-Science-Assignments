# Section A: Data Structures & Control Structures

## 1. List operations

```python
nlist = [11, 42, 74, 23, 89, 42]     # list creation
```

```python
nlist.append(100)    # append new element to the last position of list
nlist
```

```python
nlist.pop()     # remove last element from list
nlist
```

```python
nlist.insert(3, 200)    # insert new element at the specified location
nlist
```

```python
nlist.remove(74)     # remove the specified element from the list
nlist
```

```python
nlist.extend((42,36))     # extend list with given elements
nlist
```

```python
nlist.index(42)     # get the  index of given element
```

```python
c = nlist.count(42)    # count the no of ocuurence of given element
c
```

```python
nlist.sort()     # sort list in ascending order
nlist
```

```python
nlist.reverse()     # reverse the list 
nlist
```

## 4. dictionary operations

```python
person = {
    'name' : 'lisha',
    'age'  : 45,
    'city' : 'Mumbai'
}         # dictionary creation
person
```

## 5. eligibility for voting

```python
person_age = int(input("Enter Age:"))    # enter input

if(person_age >= 18):      # check condition
    print("Person is eligible to vote")
else:
    print("Person is not eligible to vote")
```

## 6. grade based on the marks

```python
marks = int(input("Enter marks:"))     # enter marks

if marks >= 90:
    print('A grade')
elif marks >= 75:
    print('B grade')
elif marks >= 50:
    print('C grade')
else:
    print('Fail')
```

## 7. positive or negative or zero or even

```python
number = int(input("Enter number:"))

if number%2 == 0:
    print('number is even')
    
if number == 0:
    print('Number is zero')
elif number < 0:
    print('Number is negative')
else:
    print('Number is positive')
```

# Section C: Exploring Pandas

## DataFrame

```python
employee = {
    'Name' : ['hina', 'mohit', 'anjali', 'priya', 'swati', 'reshu', 'joe', 'shital', 'chetan', 'shital'],
    'Age'  : [89, 65, 16, 73, 19, 46, 25, 20, 45, 20],
    'City' : ['Numbai', 'Hydrabad', 'Kolkata', 'Ambala', 'Delhi', 'Amritsar', 'Puri', 'Vizac', 'Madras', 'Vizac'],
    'Salary':[34000, 45000, 68000, None , 95000, 34000, 62000, 31000, 20000, 31000]
}     # create dictionary

employee_DF = pd.DataFrame(employee)    # create data frame with dictionary
employee_DF
```

```python
employee_DF.info()    # get structural information of dataframe
```

```python
employee_DF.describe()    # get summary of data
```

```python
employee_DF['Name']    # get feture data from Name 
```

```python
employee_DF['City']
```

```python
employee_DF.drop(columns=['City'])    # drop feature City
```

```python
employee_DF.isnull().sum()     # check whether null in data
```

```python
import warnings
warnings.filterwarnings('ignore')     # filter warnings

mean = employee_DF['Salary'].mean()    # calculate salary average
employee_DF['Salary'].fillna(mean, inplace=True)    # replace null with average value
employee_DF
```

```python
employee_DF.duplicated().sum()    # check whether any duplicates in data
```

```python
employee_DF.drop_duplicates()     # drop duplicates if any
```
