### Program that checks whether number is prime or not

```python
number = input("Enter a number: ")
prime_flag = False
if(number.isnumeric()):
    number = int(number)
    if(number<1):
        print("Number is must be greater than 1")
    else:
        i = 2;
        while(i<=number):        
            if(number%i==0):
                if(number==i):
                    print("Number is prime")
                    prime_flag = True
                break;                                     
            i=i+1       
        if(prime_flag==False):
            print("Number is not prime")
else:
    print("Number is not valid")
    
```

### Program that checks the product of random numbers is correct or not

```python
import random
number1 = random.random()
number2 = random.random()

print("Please enter the product of ",number1," and ",number2,":")
product = float(input())

answer = number1*number2

if(product==answer):
    print("The answer is correct")
else:
    print("Wrong answer")
```

### Program that prints the square of even numbers within the range

```python
def square_of_even_numbers():
    print("Square of even numbers within the range 100 to 200: ")
    for i in range(100,201):
        if(i%2==0):
            print(i ,":", i**2)

square_of_even_numbers()
```

### Program to check whether string is palindrom

```python
input_string = input("Enter String to check palindrom:")
input_string = input_string.lower()
clean_string = re.sub(r'[^\w\s]', '', input_string)

#reversed_string = clean_string[::-1]
def reverse(clean_string):
    counter= 1
    reversed_string = ''
    while(len(clean_string)>=counter):    
        reversed_string =  reversed_string + clean_string[-counter]    
        counter = counter + 1
    return reversed_string

if(clean_string == reverse(clean_string)):
    print("True")
else:
    print("False")
```
