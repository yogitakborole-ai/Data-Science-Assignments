```python
import pandas as pd    # import libraries
import numpy as np
```

```python
data = pd.read_csv("datasets/sales_data_with_discounts.csv")   # read data

data.info()
```

```python
data.head(3)
```

### <i> Discriptive Analytics for Numerical Columns

```python
num_cols = data.select_dtypes(include=['int64','float64']).columns   # select numerical data types
data[num_cols].info()
#num_cols.size
```

#### <u><i>mean</u>: it is the average of range of values; summation of all values divide by number of values
####           $ \text{mean} = \frac{\sum \text{Xi}}{\text{n}} \$


```python
mean = data[num_cols].mean()    # calculate mean
mean
```

#### <u><i>median</u>: it is the middle most value of the ordered list

```python
median = data[num_cols].median()  # calculate median
median
```

#### <u><i>mode</u>: It is most frequently occuring data

```python
mode = data[num_cols].mode()    # calculate mode
mode   
```

####  <u><i>standard deviation</u> : it is the square root of variance 
$ \sigma = \sqrt{variance} \$

```python
data["Volume"].std()    # standard deviation
```

##### We computed basic statistical measures such as mean, median, mode, and standard deviation for numerical columns in the dataset.

### <i> Data Vizualization

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

#### <u>Histograms</u>:

```python
for i in num_cols:   
    sb.displot(data[i],
            kind='hist',
            height=3,
            aspect=2,
               kde=True
           )
    plt.show()
```

#### <u>BoxPlot</u>:

```python
#for col in num_cols:
sb.boxplot(data=data[num_cols])
plt.xticks(rotation=45)
plt.show()
```

```python
desc_cols = data.select_dtypes(include=["object"]).columns
desc_cols
```

```python
for col in desc_cols:
    sb.countplot(x=data[col])
    plt.xticks(rotation=45)
    plt.show()
```

#### Histograms helped us visualize the distribution of numerical variables, revealing information about skewness and the presence of outliers. 
#### Boxplots provided a clear view of outliers and the interquartile range of numerical variables. 
#### Bar charts displayed the frequency or count of categories in categorical columns, enabling us to analyze the distribution of categories.
