```python
anime1 = anime.pivot(index='anime_id',
                                 columns='genre',
                                 values='rating')
anime1
```

```python
anime1.fillna(0, inplace=True)
```

```python
anime1
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

user_similarity = 1 - pairwise_distances(anime1.values, metric='cosine')  # similarity
```

```python
np.fill_diagonal(user_similarity, 0)
```

### Q1. Can you explain the difference between user-based and item-based collaborative filtering?

<i>Collaborative filtering</i> is a technique in recommendation system to suggest item based on the preferences and behaviours of similar users.  
<i>User-based filtering:</i> it finds other users which are similar to target users and recommends items they liked.  
<i>Item-based filtering:</i> it finds items similar to those  the user  liked and recommends them.
