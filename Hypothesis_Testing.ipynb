```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as stats
```

### given data:
$ \bar X = 3050 $      [sample mean weekly cost] 

$\mu =  1000+5*600 $   [theorietical mean weekly cost according to the cost model (W=$1,000 + $5X for X=600 units)]

$ \sigma = 5*25$   [standard deviation of population]

$\alpha = 0.05$   [significance level]

$ n=25 $    [sample size]



$ H_0: \mu <= 4000 $

$ H_1: \mu > 4000 $


```python
sample_mean = 3050
theoretical_mean = 1000+5*600
std_dev = 5*25
alpha = 0.05
sample_size = 25
```

```python
z_score = (sample_mean - theoretical_mean)/(std_dev / np.sqrt(sample_size))
p_value = 1 - stats.norm.cdf(z_score) 
p_value
```

```python
if (p_value < alpha):
    print("Reject the null hypothesis: There is a strong evedence to suppoert restaurant owners' claim that the weekly operating costs are higher than the model suggest")
else:
    print("Failed to reject the null hypothesis: There is not enough evedence to suppoert restaurant owners' claim that the weekly operating costs are higher than the model suggest")
```

# Confidence Intervals

```python
print_heads = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
import numpy as np

n = len(print_heads)
mean = np.mean(print_heads)
std_dev = np.std(print_heads)
print(f' mean of print heads is {np.round(mean,2)} \n standard deviation of print head is {np.round(std_dev,2)}')
```

```python
# Build 99% Confidence Interval Using Sample Standard Deviation

import scipy.stats as stats

t_critical_99 = stats.t.interval(0.99, loc=mean, df=(n-1), scale= std_dev/np.sqrt(n))
print(np.round(t_critical_99,2))
```

this shows, true mean lies within this interval

```python
# Build 99% Confidence Interval Using Known Population Standard Deviation

std_dev = 0.2  # known population standard deviation
z_critical_99 = stats.norm.interval(0.99, loc=mean, scale= std_dev/np.sqrt(n))
print(np.round(z_critical_99,2))
```
